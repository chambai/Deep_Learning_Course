#Explore existing DNN models in Keras
The pre-existing model names in Keras can be found here `https://keras.io/api/applications/`

image.png

```
import tensorflow as tf

# to load the model:
model = tf.keras.applications.VGG16(
    weights=None,   # loads the model with randomly initialised weights
    include_top=True,   # include_top means that a fully connected layer will be added at the end of the model.   
    classes=10,         # include_top means that you can specify this classes parameter, specifying the number of classes of your data
    input_shape=(32,32,1) # change the input shape of the network to match the dataset
)

# display the summary so we can see the architecture of the model
model.summary()
```

## Display the sumamry of other pre-existing DNNs in Keras.  For instance, try one of the EfficientNet models.  What do you notice?

```
import tensorflow as tf

# to load the model:
model = tf.keras.applications.EfficientNetB7(
    weights=None,   # loads the model with randomly initialised weights
    include_top=True,   # include_top means that a fully connected layer will be added at the end of the model.   
    classes=10,         # include_top means that you can specify this classes parameter, specifying the number of classes of your data
    input_shape=(32,32,1) # change the input shape of the network to match the dataset
)

# display the summary so we can see the architecture of the model
model.summary()
```
